```python
import pandas as pd
import numpy as np
import csv
import os
from geopy import distance
import random
import time
import ee
ee.Initialize()
```

```python
compositeToUse = ee.Image('projects/crowtherlab/nina/20200728_composite_subset')
scaleToUse = compositeToUse.projection().nominalScale().getInfo()
```

```python
treemap_data = ee.FeatureCollection('users/ninavantiel/treemap_sdm_ee/treemap_data')
```

```python
species_list = treemap_data.aggregate_array('species').distinct().getInfo()
```

```python
for species in species_list:
    print(species)
    
    try:
        testSpecies = ee.FeatureCollection('users/ninavantiel/treemap_sdm_ee/presences_sampled/sampledPoints_' + species.replace(' ', '_')).limit(1).getInfo()
    except:
        print('Sample the species.')
        
        collToSample = treemap_data.filterMetadata('species', 'equals', species)
        print(collToSample.size().getInfo())
        
        sampledPoints = collToSample.map(lambda f: f.set(
            compositeToUse.reduceRegion(
                reducer = 'first',
                geometry = f.geometry(),
                scale = scaleToUse)))
    
        task = ee.batch.Export.table.toAsset(
            collection = sampledPoints,
            description = 'sampledPoints_' + species.replace(' ', '_'),
            assetId = 'users/ninavantiel/treemap_sdm_ee/presences_sampled/sampledPoints_' + species.replace(' ', '_'))
        task.start()
    
    else:
        print('Species was already sampled.')
```
