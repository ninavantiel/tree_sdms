```python
n = 1000
cand_round1 = samples.sample(n)
cand_idx = cand_round1.index
```

```python
dist_mat = np.empty((n,n))
dist_mat[:] = 0

for i in range(n):
    cand_i = cand_round1.loc[cand_idx[i]]
    coords_i = (cand_i.lat, cand_i.lon)
    
    for j in range(i+1, n):
        cand_j = cand_round1.loc[cand_idx[j]]
        coords_j = (cand_j.lat, cand_j.lon)
        
        dist_mat[i,j] = distance.geodesic(coords_i, coords_j).meters

# make distance matrix symmetric
dist_mat = dist_mat + dist_mat.T - np.diag(dist_mat.diagonal())
```

```python
bool_mat = dist_mat < limdist
np.fill_diagonal(bool_mat, False)
```

```python
bool_mat.shape
```

```python
remove_idx = []
bool_sum = np.sum(bool_mat, axis=1)

while not np.all(bool_sum == 0):  
    to_rm = random.choice(np.where(bool_sum == max(bool_sum))[0])
    
    bool_mat = np.delete(bool_mat, to_rm, 0)
    bool_mat = np.delete(bool_mat, to_rm, 1)
    remove_idx.append(to_rm)
        
    bool_sum = np.sum(bool_mat, axis = 1)
```

```python
samples_keep = cand_round1.drop(cand_idx[remove_idx])
```
