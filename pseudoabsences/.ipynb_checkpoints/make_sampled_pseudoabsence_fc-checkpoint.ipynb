```python
pseudoabsences = pd.read_csv('pseudoabsences_sampled.csv')
nPApool = pseudoabsences.shape[0]
print(str(nPApool) + ' pseudoabsences in pool.')
```

```python
# Remove points that have null values in the covariates 
pseudoabsences = pseudoabsences[pseudoabsences.Pixel_Lat != 'None']
pseudoabsences = pseudoabsences[pseudoabsences.CGIAR_Aridity_Index != 'None']
pseudoabsences = pseudoabsences[pseudoabsences.CHELSA_BIO_Annual_Mean_Temperature != 'None']
pseudoabsences = pseudoabsences[pseudoabsences.CHELSA_BIO_Precipitation_Seasonality != 'None']

nLeft = pseudoabsences.shape[0]
print(str(nPApool-nLeft) + ' observations removed because of null values in covariates.')
```

```python
# Check that there are no more null values in the covariates
ok = True
for var in covariates:
    if pseudoabsences[pseudoabsences[var] == 'None'].shape[0] != 0: 
        print('Null values still in ' + var)
        ok = False
if ok: print('No more null values in the covariates!')
```

```python
pseudoabsences = pseudoabsences[covariates]
pseudoabsences['presence'] = 0
pseudoabsences
```

```python
pseudoabsences.to_csv('pseudoabsences_fc.csv', index = False)

subprocess.run(['gsutil', 'cp', 'pseudoabsences_fc.csv', 'gs://theforgottenrealms16180339_nina_01/'])
print('Uploading regression matrix to GCSB')

while not 'pseudoabsences_fc.csv' in subprocess.run(['gsutil', 'ls', 'gs://theforgottenrealms16180339_nina_01/'], stdout=subprocess.PIPE).stdout.decode('utf-8'):
    print('Uploading...')
    time.sleep(5)
print('Upload done!')
```

```python
subprocess.run(['/usr/local/bin/earthengine', 'upload', 'table', 
                '--asset_id=users/ninavantiel/treemap_sdm_ee/pseudoabsences',
               'gs://theforgottenrealms16180339_nina_01/pseudoabsences_fc.csv'])
print('Ingesting asset into earthengine')
```

```python
done = False
while not done:
    taskList = [str(i) for i in ee.batch.Task.list()]
    subsetList = [s for s in taskList if 'pseudoabsences' in s]
    subsubList = [s for s in subsetList if any(xs in s for xs in ['RUNNING', 'READY'])]
    if len(subsubList) >= 1: print('Ingesting...')
    else: done = True
    time.sleep(10)

print('Ingestion done!')
print(subsetList)
```
